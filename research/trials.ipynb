```python
## What is the purpose of the decorator @ensure_annotations

def product(x:int, y: int) -> int:
    return x*y
```

```python
product(2, 4)
```

```python
product(2, "4")
```

```python
# To avoid the function from getting executed when incorrect type of input is passed to the fucntion we make use of ensure_annotations

from ensure import ensure_annotations

@ensure_annotations
def product(x:int, y:int) -> int:
    return x*y
```

```python
product(2, 4)
```

```python
product(2, "4")
```
